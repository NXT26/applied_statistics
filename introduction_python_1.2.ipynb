```python
def gen_list(a,b,c):
    list=[]
    for i in range(a,b,c):
        list.append(i)
    return list
```

```python
def gen_dict(Name,Age):
    person={}
    for i in range(len(Name)):
        person[Name[i]]=Age[i]
    return person
```
